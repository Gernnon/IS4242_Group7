```python
with open('Private Residential Property Transactions.json', 'r') as f:
    data = json.loads(f.read())

# Flatten data
df = pd.json_normalize(data, record_path=['Result'])
df
```

```python
from pandarallel import pandarallel
pandarallel.initialize(progress_bar=True)

# Convert x and y (SVY21 coordinates) into Latitude and Longitude
def get_lat_long(row):
    import pyproj
    xfm = pyproj.Transformer.from_crs('EPSG:3414', 'EPSG:4326')
    return xfm.transform(row['y'], row['x'])

merged['lat_long'] = merged.parallel_apply(get_lat_long, axis=1)
```

### Incorporating Nearest MRT data

```python
%pip install xlrd
```

```python
stations = pd.read_excel('Train Station Codes and Chinese Names.xls') # From LTA DataMall
stations
```

```python
# Dataframe to store MRT station names, latitude and longitude
mrt_df = pd.DataFrame()

# Iterate through station codes to search and store info
for index, row in stations.iterrows():
    station_name = row['mrt_station_english'] + ' mrt station (' + row['stn_code'] + ')'
    station_name = station_name.upper()
    # Common search API to search for MRT stations
    url = "https://developers.onemap.sg/commonapi/search?returnGeom=Y&getAddrDetails=N&searchVal="+station_name
    response = requests.get(url)
    station_data = json.loads(response.text)
    # Normalize and convert to DataFrame
    row_df = pd.json_normalize(station_data, record_path=['results'])
    # Concatenate to the MRT df
    mrt_df = pd.concat([mrt_df, row_df], ignore_index=True)

mrt_df

```

```python
# Keep just the Station Name, Latitude and Longitude
mrt_df = mrt_df.rename(columns={'SEARCHVAL':'STATION'}).drop(columns=['X', 'Y', 'LONGTITUDE'])
mrt_df['LATITUDE'], mrt_df['LONGITUDE'] = mrt_df['LATITUDE'].astype(float), mrt_df['LONGITUDE'].astype(float)
mrt_df
```

```python
%pip install geopy
```

```python
from pandarallel import pandarallel
pandarallel.initialize(progress_bar=True)

# Map nearest MRT to each property
def nearest_mrt(row, stations=mrt_df):
    from geopy.distance import geodesic

    min_distance, nearest_stn, nearest_stn_lat_long = 10000, "", ""
    for stn, stn_lat, stn_long in zip(stations['STATION'], stations['LATITUDE'], stations['LONGITUDE']):
        distance = geodesic(row['lat_long'], (stn_lat, stn_long)).km
        if distance < min_distance:
            min_distance = distance
            nearest_stn = stn
            nearest_stn_lat_long = (stn_lat, stn_long)
    return nearest_stn, min_distance, nearest_stn_lat_long

merged['nearest_mrt'], merged['distance_from_nearest_mrt'], merged['nearest_mrt_lat_long'] = zip(*merged.parallel_apply(nearest_mrt, axis=1))
```
