```python
import pandas as pd
import json
import pyproj
import requests
import ast

from pandarallel import pandarallel
from geopy.distance import geodesic
```

```python
with open('Private Residential Property Transactions.json', 'r') as f:
    data = json.loads(f.read())

# Flatten data
property = pd.json_normalize(data, record_path=['Result'])
property
```

```python
property_fixed = pd.merge(property, property2, left_index=True, right_index=True)
property_fixed
```

## Using Functions from lucas to turn the property column X Y to Lat Long

```python
property_fixed = property_fixed.drop(columns=['transaction', 'nettPrice'])
property_fixed
```

```python
pandarallel.initialize(progress_bar=True)

# Convert x and y (SVY21 coordinates) into Latitude and Longitude
def get_lat_long(row):
    import pyproj
    xfm = pyproj.Transformer.from_crs('EPSG:3414', 'EPSG:4326')
    return xfm.transform(row['y'], row['x'])

property_fixed['lat_long'] = property_fixed.parallel_apply(get_lat_long, axis=1)
```

## Incorporating BusStops and TaxiStands data

```python
properties = pd.read_csv('Private Residential Property Transactions with MRT Data.csv')
properties
```

```python
for index, row in properties.iterrows():
    lat_long = row['lat_long']
    lat, long = ast.literal_eval(lat_long)  # convert the string to a tuple of two objects
    properties.at[index, 'lat_long'] = lat, long
```

```python
pandarallel.initialize(progress_bar=True)

# Map nearest bus_stops to each property
def nearest_bus_stop(row, stops=bus_stops):
    from geopy.distance import geodesic

    min_distance, nearest_bus_stop, nearest_bus_stop_lat_long = 10000, "", ""
    for bus_stop, bus_stop_lat, bus_stop_long in zip(stops['Description'], stops['Latitude'], stops['Longitude']):
        distance = geodesic(row['lat_long'], (bus_stop_lat, bus_stop_long)).km
        if distance < min_distance:
            min_distance = distance
            nearest_bus_stop = bus_stop
            nearest_bus_stop_lat_long = (bus_stop_lat, bus_stop_long)
    return nearest_bus_stop, min_distance, nearest_bus_stop_lat_long

properties['nearest_bus_stop'], properties['distance_from_nearest_bus_stop'], properties['nearest_bus_stop_lat_long'] = zip(*properties.parallel_apply(nearest_bus_stop, axis=1))
```
