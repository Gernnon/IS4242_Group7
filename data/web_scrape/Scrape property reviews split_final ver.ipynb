```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import pandas as pd
import re
import time
```

### Scraping Ratings for Apartments and Condominiums (Majority)

```python
df = pd.read_csv('Transactions with MRT, Bus, Taxi, Crime.csv')
ap_condo = ['Apartment', 'Condominium']
df_apart_condo = df[df['propertyType'].isin(['Apartment', 'Condominium'])].copy()
df_apart_condo['propertyName'] = df_apart_condo['project'] + ' ' + df_apart_condo['street']

ratings_df = pd.DataFrame()
ratings_df['propertyName'] = df_apart_condo.propertyName.unique()

batch1 = ratings_df.iloc[:36].copy()
batch2 = ratings_df.iloc[36:72].copy()
batch3 = ratings_df.iloc[72:108].copy()
batch4 = ratings_df.iloc[91:126].copy()
batch5 = ratings_df.iloc[126:161].copy()
batch6 = ratings_df.iloc[161:196].copy()
batch7 = ratings_df.iloc[196:].copy()
```

```python
# Initialise driver and query url
service = Service(ChromeDriverManager().install())
options = Options()
options.add_argument("--incognito")
driver = webdriver.Chrome(service=service, options=options)
url = 'https://www.google.com/search?q=99.co+'
# Count to sleep for 1 second every 5 searches
count = 0

# List to store ratings
ratings = []

# Scrape Google search results for Property Name
for property in batch4.propertyName.tolist():
    property_name = property.replace(' ', '+')
    driver.get(url + property_name)
    try:
        rating_element = driver.find_element(By.CSS_SELECTOR, "div[class^='fG8Fp uo4vr']").get_attribute('innerText')
        rating = float(re.search(r'\d+\.\d+', rating_element).group(0))
    except:
        rating = None
    ratings.append(rating)
    print(rating)
    # Sleep for 0.6 second every 5 searches
    if count % 5 == 0:
        time.sleep(0.6)
    count += 1
driver.close()

batch4['rating'] = ratings 
```

### Scraping Ratings for Landed/ Semi-Detached Properties

```python
df = pd.read_csv('Transactions with MRT, Bus, Taxi, Crime.csv')
df_landed = df[~df['propertyType'].isin(['Apartment', 'Condominium'])].copy()
df_landed['propertyName'] = df_landed['project'] + ' ' + df_landed['street']

ratings_df2 = pd.DataFrame()
ratings_df2['propertyName'] = df_landed.propertyName.unique()

batch8 = ratings_df2[:30].copy()
batch9 = ratings_df2[30:60].copy()
batch10 = ratings_df2[60:].copy()
```

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options

# Initialise driver and query url
service = Service(ChromeDriverManager().install())
options = Options()
options.add_argument("--incognito")
driver = webdriver.Chrome(service=service, options=options)
url = 'https://www.google.com/search?q=99.co+'
# Count to sleep for 1 second every 5 searches
count = 0

# List to store ratings
ratings = []

# Scrape Google search results for Property Name
for property in batch10.propertyName.tolist():
    property_name = property.replace(' ', '+')
    driver.get(url + property_name)
    try:
        rating_element = driver.find_element(By.CSS_SELECTOR, "div[class^='fG8Fp uo4vr']").get_attribute('innerText')
        rating = float(re.search(r'\d+\.\d+', rating_element).group(0))
    except:
        rating = None
    ratings.append(rating)
    print(rating)
    # Sleep for 0.6 second every 5 searches
    if count % 5 == 0:
        time.sleep(0.6)
    count += 1
driver.close()

batch10['rating'] = ratings 
```

```python
# Open csv as df and merge
batch1_df = pd.read_csv('ratings_batch1.csv')
batch2_df = pd.read_csv('ratings_batch2.csv')
batch3_df = pd.read_csv('ratings_batch3.csv')
batch4_df = pd.read_csv('ratings_batch4.csv')
batch5_df = pd.read_csv('ratings_batch5.csv')
batch6_df = pd.read_csv('ratings_batch6.csv')
batch7_df = pd.read_csv('ratings_batch7.csv')
batch8_df = pd.read_csv('ratings_batch8.csv')
batch9_df = pd.read_csv('ratings_batch9.csv')
batch10_df = pd.read_csv('ratings_batch10.csv')

combined_ratings = pd.concat([batch1_df, batch2_df, batch3_df, batch4_df, batch5_df, batch6_df, batch7_df, batch8_df, batch9_df, batch10_df], ignore_index=True)
combined_ratings
```

```python
# Clean up those with too large values
combined_ratings.loc[combined_ratings.rating > 5.0, 'rating'] = 4.0
# Clean up those with false 0.0 values
combined_ratings.loc[combined_ratings.rating == 0.0, 'rating'] = 4.0
# Clean up NaN values
combined_ratings.loc[combined_ratings.rating.isna(), 'rating'] = 4.0 # Most have rating of 4, else filling with median
```

### Merge Unique Property Ratings with Transaction data

```python
combined_ratings = pd.read_csv('Unique property rating.csv')
df = pd.read_csv('Transactions with MRT, Bus, Taxi, Crime.csv')
df['propertyName'] = df['project'] + ' ' + df['street']
df = pd.merge(df, combined_ratings.drop_duplicates(subset=['propertyName']), how='left', left_on='propertyName', right_on='propertyName')
df = df.drop(columns='propertyName')
df.info()
```
